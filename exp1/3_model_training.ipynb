# HAI-21.03 Dataset Model Training

This notebook focuses on building and training an LSTM autoencoder model for anomaly detection.

```python
# Import required libraries
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout, LSTM, LayerNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import precision_recall_curve
import warnings
warnings.filterwarnings('ignore')

# Set plot style
plt.style.use('ggplot')
sns.set(style="darkgrid")

# Set random seeds for reproducibility
np.random.seed(42)
tf.random.set_seed(42)
```

## 1. Global Variables Setup

```python
# Set paths
OUTPUT_DIR = '../../hai-security-dataset/processed'
MODEL_DIR = '../../models/simplified_lstm'

# Create model directory if it doesn't exist
os.makedirs(MODEL_DIR, exist_ok=True)

# Set model parameters
SEQ_LENGTH = 128  # Sequence length (time steps)
STRIDE = 10       # Sliding window stride

# Set training parameters
EPOCHS = 5
BATCH_SIZE = 32
VALIDATION_SPLIT = 0.2
PATIENCE = 5
LEARNING_RATE = 0.001

# Set anomaly detection parameters
CONTAMINATION = 0.005  # Contamination rate for dynamic threshold
```

## 2. Load Processed Data

```python
# Load processed training data
train_df = pd.read_csv(os.path.join(OUTPUT_DIR, 'train_processed.csv'))
print(f"Loaded processed training data: {train_df.shape[0]} rows, {train_df.shape[1]} columns")

# Load feature information
with open(os.path.join(OUTPUT_DIR, 'feature_info.pkl'), 'rb') as f:
    feature_info = pickle.load(f)
    
top_features = feature_info['top_features']
n_components = feature_info['n_components']

print(f"Loaded feature information: {len(top_features)} top features, {n_components} PCA components")

# Load scaler and PCA
with open(os.path.join(OUTPUT_DIR, 'scaler.pkl'), 'rb') as f:
    scaler = pickle.load(f)
    
with open(os.path.join(OUTPUT_DIR, 'pca.pkl'), 'rb') as f:
    pca = pickle.load(f)
    
print("Loaded scaler and PCA")
```

```python
# Extract features
X_train_features = train_df[top_features].values

# Scale data
X_train_scaled = scaler.transform(X_train_features)

# Apply PCA
X_train_pca = pca.transform(X_train_scaled)

print(f"Prepared training data: {X_train_pca.shape}")
```

## 3. Time Series Preprocessing

```python
def create_sequences(data, seq_length=128, stride=10):
    """
    Create sequences for time series models.
    
    Args:
        data (np.array): Input data
        seq_length (int): Sequence length
        stride (int): Stride for sliding window
        
    Returns:
        np.array: Sequences
    """
    sequences = []
    for i in range(0, len(data) - seq_length + 1, stride):
        seq = data[i:i+seq_length]
        sequences.append(seq)
    
    return np.array(sequences)
```

```python
# Create sequences
X_train_seq = create_sequences(X_train_pca, SEQ_LENGTH, STRIDE)

print(f"Training sequences shape: {X_train_seq.shape}")
print(f"Number of sequences: {X_train_seq.shape[0]}")
print(f"Sequence length: {X_train_seq.shape[1]}")
print(f"Number of features: {X_train_seq.shape[2]}")
```

```python
# Visualize a few sequences
plt.figure(figsize=(15, 10))
for i in range(3):  # Show first 3 features
    plt.subplot(3, 1, i+1)
    for j in range(5):  # Show first 5 sequences
        plt.plot(X_train_seq[j, :, i], label=f'Sequence {j+1}')
    plt.title(f'Feature {i+1} Sequences')
    plt.xlabel('Time Step')
    plt.ylabel(f'Feature {i+1} Value')
    plt.legend()
plt.tight_layout()
plt.show()
```

## 4. Build LSTM Autoencoder Model

```python
def build_simplified_lstm_autoencoder(seq_length, n_features):
    """
    Build a simplified LSTM autoencoder model.
    
    Args:
        seq_length (int): Sequence length
        n_features (int): Number of features
        
    Returns:
        Model: Keras model
    """
    # Input layer
    inputs = Input(shape=(seq_length, n_features), name='input_layer')
    
    # Encoder - simplified to just one LSTM layer
    x = LSTM(16, activation='tanh', return_sequences=False)(inputs)
    x = LayerNormalization()(x)
    x = Dropout(0.2)(x)
    
    # Bottleneck
    encoded = Dense(8, activation='relu')(x)
    
    # Decoder - simplified
    x = Dense(16, activation='relu')(encoded)
    x = tf.keras.layers.RepeatVector(seq_length)(x)
    
    x = LSTM(16, activation='tanh', return_sequences=True)(x)
    x = LayerNormalization()(x)
    x = Dropout(0.2)(x)
    
    # Output layer
    outputs = tf.keras.layers.TimeDistributed(Dense(n_features))(x)
    
    # Create model
    model = Model(inputs=inputs, outputs=outputs)
    
    # Compile model
    model.compile(optimizer=Adam(learning_rate=LEARNING_RATE), loss='mse')
    
    return model
```

```python
# Build model
print("Building simplified LSTM autoencoder model...")
model = build_simplified_lstm_autoencoder(SEQ_LENGTH, X_train_pca.shape[1])

# Print model summary
model.summary()
```

```python
# Visualize model architecture
tf.keras.utils.plot_model(model, to_file=os.path.join(MODEL_DIR, 'model_architecture.png'), 
                          show_shapes=True, show_layer_names=True)
```

## 5. Train Model

```python
# Define callbacks
callbacks = [
    EarlyStopping(monitor='val_loss', patience=PATIENCE, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=0.0001),
    ModelCheckpoint(os.path.join(MODEL_DIR, 'simplified_lstm.h5'), monitor='val_loss', save_best_only=True)
]

# Train model
print("Training simplified LSTM autoencoder model...")
history = model.fit(
    X_train_seq, X_train_seq,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_split=VALIDATION_SPLIT,
    callbacks=callbacks,
    verbose=1
)
```

```python
# Plot training history
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Simplified LSTM Autoencoder Training History')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

## 6. Model Evaluation on Training Data

```python
# Get predictions on training data
X_train_pred = model.predict(X_train_seq)

# Calculate reconstruction error
train_mse = np.mean(np.square(X_train_seq - X_train_pred), axis=(1, 2))

print(f"Mean reconstruction error: {np.mean(train_mse):.6f}")
print(f"Min reconstruction error: {np.min(train_mse):.6f}")
print(f"Max reconstruction error: {np.max(train_mse):.6f}")
```

```python
# Visualize reconstruction error distribution
plt.figure(figsize=(12, 6))
plt.hist(train_mse, bins=50, alpha=0.7)
plt.title('Reconstruction Error Distribution')
plt.xlabel('Reconstruction Error (MSE)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

## 7. Determine Anomaly Detection Threshold

```python
def determine_threshold(model, X_train_seq, method='dynamic', contamination=0.01, percentile=99):
    """
    Determine threshold for anomaly detection.
    
    Args:
        model (Model): Trained model
        X_train_seq (np.array): Training sequences
        method (str): Method for threshold determination ('dynamic', 'percentile', or 'iqr')
        contamination (float): Expected proportion of anomalies (for dynamic method)
        percentile (float): Percentile for threshold (for percentile method)
        
    Returns:
        float: Threshold value
    """
    # Get predictions from model
    X_train_pred = model.predict(X_train_seq)
    train_mse = np.mean(np.square(X_train_seq - X_train_pred), axis=(1, 2))
    
    if method == 'dynamic':
        # Simulate anomalies using highest reconstruction errors
        simulated_anomalies = np.zeros(len(train_mse))
        anomaly_count = int(contamination * len(train_mse))
        anomaly_indices = np.argsort(train_mse)[-anomaly_count:]
        simulated_anomalies[anomaly_indices] = 1
        
        # Calculate precision-recall curve
        precisions, recalls, thresholds = precision_recall_curve(simulated_anomalies, train_mse)
        
        # Find threshold that maximizes F1 score
        f1_scores = 2 * (precisions * recalls) / (precisions + recalls + 1e-10)
        optimal_idx = np.argmax(f1_scores)
        threshold = thresholds[optimal_idx] if optimal_idx < len(thresholds) else thresholds[-1]
        
    elif method == 'percentile':
        # Use percentile method
        threshold = np.percentile(train_mse, percentile)
        
    elif method == 'iqr':
        # Use IQR method
        q1 = np.percentile(train_mse, 25)
        q3 = np.percentile(train_mse, 75)
        iqr = q3 - q1
        threshold = q3 + 1.5 * iqr
        
    else:
        raise ValueError(f"Unknown threshold method: {method}")
    
    return threshold
```

```python
# Determine threshold using different methods
dynamic_threshold = determine_threshold(model, X_train_seq, method='dynamic', contamination=CONTAMINATION)
percentile_threshold = determine_threshold(model, X_train_seq, method='percentile', percentile=99)
iqr_threshold = determine_threshold(model, X_train_seq, method='iqr')

print(f"Dynamic threshold: {dynamic_threshold:.6f}")
print(f"Percentile-based threshold (99th): {percentile_threshold:.6f}")
print(f"IQR-based threshold: {iqr_threshold:.6f}")
```

```python
# Visualize thresholds on reconstruction error distribution
plt.figure(figsize=(12, 6))
plt.hist(train_mse, bins=50, alpha=0.7)
plt.axvline(x=dynamic_threshold, color='r', linestyle='--', label=f'Dynamic Threshold: {dynamic_threshold:.6f}')
plt.axvline(x=percentile_threshold, color='g', linestyle='--', label=f'Percentile Threshold: {percentile_threshold:.6f}')
plt.axvline(x=iqr_threshold, color='b', linestyle='--', label=f'IQR Threshold: {iqr_threshold:.6f}')
plt.title('Reconstruction Error Distribution with Thresholds')
plt.xlabel('Reconstruction Error (MSE)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

## 8. Save Model and Threshold

```python
# Save threshold
threshold = dynamic_threshold  # Use dynamic threshold as default
with open(os.path.join(MODEL_DIR, 'threshold.txt'), 'w') as f:
    f.write(str(threshold))

# Save model parameters
model_params = {
    'seq_length': SEQ_LENGTH,
    'stride': STRIDE,
    'n_components': n_components,
    'threshold': threshold,
    'top_features': top_features
}

with open(os.path.join(MODEL_DIR, 'model_params.pkl'), 'wb') as f:
    pickle.dump(model_params, f)

print(f"Saved model and threshold to {MODEL_DIR}")
```
